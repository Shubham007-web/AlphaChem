```python
# Function to extract and aggregate lesson data under a specific chapter
def aggregate_lesson_features(chapter_name, data):
    """
    Aggregates lesson features (Vocabulary, Essential Questions, Learning Objectives) under a specific chapter.

    Parameters:
    - chapter_name (str): The name of the chapter.
    - data (DataFrame): The full dataset from the CSV file.

    Returns:
    - dict: Aggregated lesson features for the chapter.
    """
    # Filter data for the specific chapter
    chapter_data = data[data['Chapter'] == chapter_name]

    # Aggregating lesson-specific features for the chapter
    lessons = chapter_data['Lesson'].unique()
    vocabulary = chapter_data['Vocabulary'].unique()
    essential_questions = chapter_data['Essential_Question'].unique()
    learning_objectives = chapter_data['Learning_Objective'].unique()
    
    # Combine all aggregated data into a dictionary
    aggregated_features = {
        'lessons': lessons,
        'vocabulary': vocabulary,
        'essential_questions': essential_questions,
        'learning_objectives': learning_objectives
    }
    
    return aggregated_features
```

```python
# Function to extract chapter data from the CSV
def get_chapter_data(chapter_name, data):
    chapter_data = data[data['Chapter'] == chapter_name].iloc[0]
    return chapter_data
```

```python
# Chapter Opener prompt with aggregated lesson features
def chapter_opener_prompt_aggregated(unit_name, unit_phenomenon, chapter_name, chapter_phenomenon, aggregated_features):
    """
    Generates a precise Chapter Opener prompt with aggregated lesson features under the chapter.

    Parameters:
    - unit_name (str): The name of the unit.
    - unit_phenomenon (str): The overarching phenomenon or storyline of the unit.
    - chapter_name (str): The name of the chapter.
    - chapter_phenomenon (str): The phenomenon or storyline specific to the chapter.
    - aggregated_features (dict): Aggregated features of lessons under the chapter (vocabulary, essential questions, learning objectives).

    Returns:
    - str: The generated Chapter Opener prompt.
    """
    # Extract the aggregated features
    vocabulary = ", ".join(aggregated_features['vocabulary'])
    essential_questions = "\n- ".join(aggregated_features['essential_questions'])
    learning_objectives = "\n- ".join(aggregated_features['learning_objectives'])
    lessons_overview = ", ".join(aggregated_features['lessons'])

    # Deriving the Big Idea from the aggregated learning objectives
    big_idea = f"The central concept of this chapter is derived from these learning objectives:\n- {learning_objectives}"

    # Deriving STEM Task from the chapter phenomenon
    chapter_stem_task = f"How does the phenomenon '{chapter_phenomenon}' impact the key concepts explored in {chapter_name}? Students will investigate this through various tasks."

    return f"""
    # Chapter Opener: {chapter_name} in {unit_name}

    ## Essential Questions:
    - The essential questions guiding this chapter are:
      - {essential_questions}
    - These questions encourage critical thinking and are connected to the unit’s overarching phenomenon: "{unit_phenomenon}."

    ## Big Idea:
    - The big idea for this chapter:
      - {big_idea}

    ## Phenomenon-Based Learning:
    - The chapter focuses on the phenomenon:
      - "{chapter_phenomenon}"
    - Students will explore this phenomenon in the context of real-world applications and scientific investigations.

    ## Chapter STEM Task:
    - By the end of the chapter, students will complete a STEM Task, focusing on:
      - {chapter_stem_task}
    - This task will connect lessons to practical applications and encourage scientific inquiry.

    ## Key Vocabulary:
    - Throughout this chapter, students will use the following vocabulary terms:
      - {vocabulary}

    ## Lessons Overview:
    - This chapter includes the following lessons:
      - {lessons_overview}

    ## Introduction:
    - Begin by introducing the chapter phenomenon and vocabulary, and encourage students to think critically about the essential questions.
    """

# Chapter Wrap-Up prompt with aggregated lesson features
def chapter_wrapup_prompt_aggregated(chapter_name, chapter_phenomenon, aggregated_features, performance_expectations, disciplinary_core_ideas):
    """
    Generates a precise Chapter Wrap-Up prompt based on aggregated lesson features under the chapter.

    Parameters:
    - chapter_name (str): The name of the chapter.
    - chapter_phenomenon (str): Revisiting the phenomenon introduced in the chapter.
    - aggregated_features (dict): Aggregated features of lessons under the chapter (vocabulary, essential questions, learning objectives).
    - performance_expectations (str): Performance expectations for the chapter.
    - disciplinary_core_ideas (str): Disciplinary Core Ideas that the chapter aligns with.

    Returns:
    - str: The generated Chapter Wrap-Up prompt.
    """
    # Extracting aggregated learning objectives
    learning_objectives = "\n- ".join(aggregated_features['learning_objectives'])
    
    # Main ideas derived from the learning objectives and disciplinary core ideas
    main_ideas = f"In this chapter, students explored the following key ideas:\n- {learning_objectives}\n- These align with the disciplinary core concepts: {disciplinary_core_ideas}."

    # Revisiting the phenomenon and reflecting on understanding
    revisiting_phenomenon = f"Now that we have completed the lessons, let’s revisit the chapter phenomenon: '{chapter_phenomenon}' and reflect on how our understanding has evolved."

    # Bring It Together section to synthesize learning
    bring_it_together = f"The lessons in this chapter helped students build a comprehensive understanding of '{chapter_phenomenon}' and how it relates to {disciplinary_core_ideas}."

    # Reflection prompts for students
    reflective_journal = f"Reflect on how your understanding of {learning_objectives} has developed throughout the chapter. How has the phenomenon '{chapter_phenomenon}' shaped your thinking?"

    # Formative assessment questions based on performance expectations
    formative_assessment = f"To assess your understanding, answer the following questions related to {performance_expectations}. These questions will include both short and long responses."

    return f"""
    # Chapter Wrap-Up: {chapter_name}

    ## Summary of Main Ideas:
    - The main concepts explored in this chapter include:
      - {main_ideas}
    - These ideas were critical for understanding the chapter phenomenon and its connection to real-world applications.

    ## Revisiting the Phenomenon:
    - Let’s revisit the chapter phenomenon:
      - "{revisiting_phenomenon}"
    - Encourage students to think critically about how their understanding of the phenomenon has changed through the lessons.

    ## Bring It Together:
    - The "Bring It Together" section synthesizes the learning:
      - {bring_it_together} and A small paragraph showing students how the mini tasks in the lessons fit into the Chapter STEM project; 
      in essence, Bring It Together! synthesizes learning from across the chapter. Be specific in how the lessons’ mini- STEM tasks contribute to the larger chapter project. 
      Ensure the learning progression along the chapter integrates knowledge from previous lessons, reinforcing the connection to the overall unit phenomenon.


    ## Reflective Journal:
    - Ask students to reflect on their learning:
      - {reflective_journal}

    ## Formative Assessment:
    - To assess your understanding, complete the following formative assessment:
      - {formative_assessment}
    """

```

```python
# Generate Chapter Opener and Wrap-Up prompts from CSV data
def generate_chapter_prompts(chapter_name, data):
    # Aggregating lesson-specific features under the chapter
    aggregated_features = aggregate_lesson_features(chapter_name, data)

    # Extracting additional chapter-level fields
    chapter_data = get_chapter_data(chapter_name, data)
    unit_name = chapter_data['Unit']
    unit_phenomenon = chapter_data['Unit_Phenomena(Storyline)']
    chapter_phenomenon = chapter_data['Chapter_Phenomena(Storyline)']
    performance_expectations = chapter_data['Performance_Expectations']
    disciplinary_core_ideas = chapter_data['Disciplinary_Core_Ideas']

    # Generate Chapter Opener
    chapter_opener = chapter_opener_prompt_aggregated(unit_name, unit_phenomenon, chapter_name, chapter_phenomenon, aggregated_features)

    # Generate Chapter Wrap-Up
    chapter_wrapup = chapter_wrapup_prompt_aggregated(chapter_name, chapter_phenomenon, aggregated_features, performance_expectations, disciplinary_core_ideas)

    return chapter_opener, chapter_wrapup

```

```python
# Example: Select a chapter from the CSV file
chapter_list = data['Chapter'].unique()
selected_chapter = 'Chapter 5: The Periodic Table and Chemical Trends'  # Example chapter

# Generate the Chapter Opener and Wrap-Up prompts
chapter_opener, chapter_wrapup = generate_chapter_prompts(selected_chapter, data)

# Display the generated prompts
print("Chapter Opener Prompt:\n", chapter_opener)
print("\nChapter Wrap-Up Prompt:\n", chapter_wrapup)
```
