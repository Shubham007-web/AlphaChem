```python
# Prompt 2.2 - SB Lesson Structure
def lesson_prompt_template(unit_name, chapter_name, lesson_name,  Essential_Question, lesson_vocabulary, lesson_objective, phenomenon):
    return f"""
    Generate comprehensive content for Lesson "{lesson_name}" under Chapter "{chapter_name}" of Unit "{unit_name}".
    The lesson should follow the structure below, aligning with the **Big Idea** and driving the essential learning outcomes.

    ### Lesson Title
    Given by S&S

    ### Big Idea:
    - One line, which addresses the main concept of the lesson.
    - A subordinate of the Chapter's Big Idea that addresses the main concepts in the lesson.

    ### Essential Questions:
    Adjust the following Essential Questions (EQs) to ensure they drive curiosity and critical thinking:
    - Up to 3 inquiry-based, open-ended questions related to the chapter's phenomenon. (E.g., “Why is water a liquid if it is such a small and light molecule?”).
    - {Essential_Question}
    - These questions should spark curiosity, promote critical thinking, and encourage further exploration.

    ### Phenomenon-Based Learning:
    The lesson should build upon the chapter's storyline and introduce a specific aspect, question, or issue that will be explored through hands-on tasks. Focus on connecting the phenomenon to the lesson's tasks and investigations.
    - Phenomenon: {phenomenon}

    ### Vocabulary:
    Provide definitions for the key terms in the lesson. Ensure students understand these terms, as they will be essential to grasping the main concepts.
    - {lesson_vocabulary}

    ### SMART Lesson Objectives:
    The following are 3-4 Specific, Measurable, Achievable, Relevant, and Time-based (SMART) learning objectives:
    - {lesson_objective}
    - These should target both content knowledge and skills, with progressively higher levels of complexity aligned to Bloom’s Taxonomy.
    - Avoid vague verbs like "understand" or "know". Use measurable action verbs such as "apply," "analyze," or "create."

    ### Ignite (Engage) [Approx. 500 words]:
    - Introduce the lesson with a phenomenon-related question or simple task that grabs students’ attention.
    - The task does not need to be rigorous or assessed but should be inclusive for all students (consider students with impairments). For example: “What happens if we cool down a liquid below its freezing point?”
    - Include the introduction of AI at least once per chapter through an inquiry-based task to encourage students to ask questions about the lesson's topic.

    ### Direct Instruction (Pre-Explore Section) [Approx. 500 words]:
    - Provide background information to build student understanding of the topic before moving to hands-on activities.
    - Avoid long passages of text without interactive elements (e.g., asking students to take notes, discuss observations, or answer scaffolded questions).
    - Relate content to the phenomenon, extracting examples to strengthen the storyline.

    ### Progress Check (Evaluate) [Approx. 200 words]:
    - Include up to 3 scaffolded questions (DOK Levels 1, 2, and 3) to ensure students can connect procedures to the content in the "Explore" section.
    - These should help students reflect on the lesson so far and guide them towards the next step.

    ### Pathfinder (Explore) [Hands-On Section, Approx. 1000 words]:
    - Provide step-by-step instructions for the hands-on activity, ensuring students collect and analyze data.
    - Include at least 1 virtual lab per chapter.
    - Encourage students to engage in group discussions and collaborative learning.
    - Use prompts that guide critical analysis of data and help students draw conclusions connecting the inquiry to the phenomenon.
    - Incorporate **Retrieval Practice** by adding questions or quizzes during or after the activity to ensure knowledge recall from previous lessons.

    ### Lightbulb (Explain) [Minds-On Section, Approx. 3000 words]:
    - Please remember this consist main content of the lesson and the content should be agline with stroyline too.
     - Provide a structured and detailed explanation of the key concepts related to the lesson.
     - Ensure coherence with the chapter's content and key objectives.
     - Include relevant examples, case studies, or real-world applications to illustrate the concepts.
     - Break down complex ideas into easily digestible sections or subsections.
     - Incorporate any relevant formulas, algorithms, or theoretical foundations.
     - If applicable, discuss any ethical considerations or potential impacts of the topic.
     - Include at least one visual aid (e.g., a diagram, flowchart, or table) to support the explanation.
    - Provide prompts that help students make sense of the hands-on activity by applying prior knowledge and discussing results.
    - Expand on the key concepts explored in the "Explore" section, linking back to the lesson's SMART objectives.
    - For every major concept, include one solved example problem followed by a question for students to solve as a Progress Check.
    - Reinforce student understanding with discussions or guided problem-solving.

    ### Progress Check (Evaluate) [Approx. 300 words]:
    - Pose up to 3 scaffolded questions (DOK Levels 1, 2, and 3) related to the Explain section to confirm understanding of the main concepts.

    ### Power Up (Elaborate) [Approx. 500 words]:
    - Introduce mini-tasks or open-ended questions that encourage students to think critically and relate lesson content to the chapter phenomenon.
    - Include opportunities for students to ask further questions, using the Questioneer to expand their understanding.

    ### Progress Check (Final Evaluation) [Approx. 800 words]:
    * Frame 1 detailed debate questions that encourage critical thinking and application of the learned concepts in the lesson. The Question should have:
    - A clear prompt
    - Potential arguments for multiple viewpoints
    - Suggested points for discussion
    * Provide 8 assessment questions:
    - 4 multiple-choice questions with 4 options each, including the correct answer and plausible distractors
    - 4 long-answer questions that require application of knowledge and higher-order thinking skills
    The alignment of questions must be uniform. MCQs and LA questions should be placed properly like all MCQs are placed first and then followed by LA Questions. More questions can be added.

    - Ensure all questions align with the unit learning outcomes and the content provided in the "Engage" and "Explain" sections.
    
    ### Beyond the Lesson (Extend) [Optional, Approx. 400 words]:
    - Introduce additional tasks, readings, or challenges that encourage students to apply their knowledge to real-world problems.
    - Provide opportunities for spaced practice, reinforcing key concepts over time to improve long-term retention.
    """
```

```python

def generate_lesson_content(unit_name, chapter_name, lesson_name, lesson_objective):
    messages = [
        {"role": "system", "content": f"You are a chemistry textbook writer for age 14-15 years, and content should be easy to read for this please use these formula {reading_score} to make easy languge to read and that should greater 90, generated in a way that feels like it's written by a human and target audience is USA school grade 9."},
        {"role": "user", "content": lesson_prompt_template(unit_name, chapter_name, lesson_name,  Essential_Question, lesson_vocabulary, lesson_objective, phenomenon)}
    ]

    # Generate the content using OpenAI's GPT model
    return generate_content_with_chat(messages)

# Function to extract numbers from unit, chapter, and lesson names for file naming
def extract_number(text):
    return re.search(r'\d+', text).group()

# Function to save the generated content to a .docx file with the specific naming format
def save_lesson_content_to_docx(unit_name, chapter_name, lesson_name, lesson_content):
    # Create a new document
    doc = Document()
    doc.add_heading(f'Unit: {unit_name}', level=1)
    doc.add_heading(f'Chapter: {chapter_name}', level=2)
    doc.add_heading(f'Lesson: {lesson_name}', level=3)

    # Add the generated content to the document
    doc.add_paragraph(lesson_content)

    # Extract unit, chapter, and lesson numbers from their names for the file name
    unit_number = extract_number(unit_name)
    chapter_number = extract_number(chapter_name)
    lesson_number = extract_number(lesson_name)

    # Construct the file name in the format U1Ch1L4.docx
    file_name = f'U{unit_number}Ch{chapter_number}L{lesson_number}.docx'

    # Save the document with the constructed file name
    doc.save(file_name)
    print(f'Content saved to {file_name}')
```

```python
# Example usage to generate and save lesson content
unit_name = "Unit 1: Foundations of Chemistry"
chapter_name = "Chapter 1: The Language of Chemistry"
lesson_name = "Lesson 4: Scientific Inquiry and Experimental Design"

# Generate detailed content for Engage, Explain, and Evaluate sections
lesson_content = generate_lesson_content(unit_name, chapter_name, lesson_name,lesson_objective)

# Print the generated content (optional)
print(lesson_content)

# Save the generated content to a .docx file in the format U1Ch1L4.docx
save_lesson_content_to_docx(unit_name, chapter_name, lesson_name, lesson_content)
```
